```python
# Import necessary libraries
import pandas as pd
from datetime import datetime
from serpapi import GoogleSearch
```

```python
# Set your SerpAPI key
serp_api_key = 'key'

# Specify the search parameters
params = {
    'engine': 'google_jobs',
    'q': 'data egineer',  # Query for data scientist jobs
    'location': 'United States',  # Location for the job search
    'api_key': serp_api_key
}

# Make a request to the SerpAPI using the GoogleSearch class
search = GoogleSearch(params)
results = search.get_dict()

# Extract the job results
jobs_results = results["jobs_results"]

# Convert the job results to a DataFrame
df = pd.DataFrame(jobs_results)
```
